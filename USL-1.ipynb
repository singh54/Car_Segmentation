### Q3.a, Q3.b, Q3.c

```python
from sklearn.cluster import KMeans

clusters = []

for i in range(1, 11):
    km = KMeans(n_clusters=i).fit(X)
    clusters.append(km.inertia_)
    print(km.inertia_)
    
fig, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax,marker='o')
ax.set_title('Searching for Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')

# Annotate arrow
ax.annotate('Possible Elbow Points', xy=(2,72921239.89189374), xytext=(4, 100000000), xycoords='data',          
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='orange', lw=2))

ax.annotate('Possible Elbow Points', xy=(5, 18811096.49806816), xytext=(4, 100000000), xycoords='data',          
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='orange', lw=2))

ax.annotate('Possible Elbow Points', xy=(3,33849580.50405829), xytext=(4, 100000000), xycoords='data',          
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='orange', lw=2))


plt.show()
```

### Q3.d, Q3.e, Q3.f

```python
# 3 cluster
km3 = KMeans(n_clusters=3).fit(X)

df['Labels_3'] = km3.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(df['wt'], df['mpg'], hue=df['Labels_3'], 
                palette=sns.color_palette('hls', 3))
plt.title('KMeans with 3 Clusters')
plt.show()
```

```python
pca=PCA(n_components=10)
pca_data=pca.fit_transform(X)
percentage_var_explained=pca.explained_variance_/np.sum(pca.explained_variance_)

cum_var_explained=np.cumsum(percentage_var_explained)

plt.figure(1,figsize=(12,8))
plt.clf()
plt.plot(cum_var_explained,linewidth=2,marker='o')
plt.plot((0,9),(0.9,0.9),'--',marker='o')
plt.grid()
plt.xlabel('n_components')
plt.ylabel('Cumulative_explained_Variance')
plt.show()
```
